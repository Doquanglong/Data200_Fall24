```python
# .agg() function

countries.agg({'lifeExpectancy':['mean','median'], 'gdpPerCapita':['median','mean']})
```

```python
# .groupby()
#countries.groupby('continent').mean(numeric_only=True)

# if you don't want continent to be row index, you can use reset_index()
countries.groupby('continent').mean(numeric_only=True).reset_index()
```

```python
# Practice
# return a table showing mean and median lifeExpectancy for each country
# Hint: combine groupby() with agg()
countries.groupby('country').agg({'lifeExpectancy':['mean','median']})
```

```python
# to drop cols or rows from a df use .drop() func. You can specifcy axis=0 if you are dropping rows, and =1 if you are
# dropping cols
# create a copy of the df first so we don't lose our work
my_copy = countries.copy()
#my_copy.drop(columns=['gdpPerCapita']) #or
my_copy.drop('gdpPerCapita',axis=1)
```

```python
# dropping rows
# create a copy of the df first so we don't lose our work
my_copy = countries.copy()
my_copy.drop(0, axis=0)
```

```python
# isin()
# filter the df to countries: 'Zimbabwe', 'Italy', 'Brazil'
# long and recommended way:
countries[(countries.country=='Zimbabwe') | (countries.country=='Italy') | \
         (countries.country=='Brazil')]

# isin() takes a list as input
mylist = ['Zimbabwe','Italy','Brazil']
countries[countries.country.isin(mylist)]
```

```python
# unique()
countries.year.unique()
```

```python
# value_counts()
countries.continent.value_counts()

# get the number of countries in each continent?
# hint: .count() can come in handy
# a second hint: drop_duplicates() that can optionally take in a column and drop rows if that col has duplicates
# approach 1: value_counts()
countries.drop_duplicates('country').continent.value_counts()
countries.drop_duplicates('country').continent.value_counts(normalize=True) # returns the proportions instead of raw counts

# approach 2: another approach using a method (?) from last class.
countries.drop_duplicates('country').groupby('continent').count()['country']

# approach 2b: also with groupby
countries.groupby(['continent','year']).count()['country'].reset_index().drop(columns=['year']).drop_duplicates()\
.reset_index(drop=True)
```

```python
# are there any countries in the data that stopped existing?
# and, are there any countries that are recently established in the data?
countries.groupby('year').size() # totals by year
mydf = countries.groupby('country').count()['year'].reset_index().rename(columns={'year':'num_year'}) # totals by country
mydf
```
